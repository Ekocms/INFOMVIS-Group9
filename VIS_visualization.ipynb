```python
print("Dataset shape: ", df.shape)   # (rows, columns)
print("\nColumn information:")
print(df.info())

print("\nFirst 5 rows:")
display(df.head())

print("\nBasic statistical summary of numerical columns:")
display(df.describe())

print("\nSummary statistics for all columns (including categorical/text):")
display(df.describe(include="all").T)


# Missing values visualization


missing_counts = df.isna().sum().sort_values(ascending=False)
missing_ratio = (missing_counts / len(df)) * 100

missing_df = pd.DataFrame({
    "missing_count": missing_counts,
    "missing_ratio": missing_ratio
})

print("\nMissing value statistics：")
display(missing_df)

top_n = 20
plt.figure(figsize=(12, 6))
sns.barplot(
    x=missing_df.index[:top_n],
    y=missing_df["missing_ratio"][:top_n]
)
plt.xticks(rotation=75, ha="right")
plt.ylabel("Missing Ratio (%)")
plt.title("Top 20 Columns with Highest Missing Ratio")
plt.tight_layout()
plt.show()

# Categorical variable distributions


print("\nList of column names:")
for i, col in enumerate(df.columns):
    print(i, col)

col_country = "Country"
col_city = "City"
col_population = "City population"
col_spatial_scale = "Spatial scale"
col_area_before = "Type of area before implementation of the NBS"
col_stage = "Present stage of the intervention"

if col_country in df.columns:
    country_counts = df[col_country].value_counts().head(15)
    plt.figure(figsize=(10, 6))
    sns.barplot(x=country_counts.values, y=country_counts.index)
    plt.xlabel("Number of NbS interventions")
    plt.ylabel("Country")
    plt.title("Top 15 Countries by Number of NbS Interventions")
    plt.tight_layout()
    plt.show()

if col_city in df.columns:
    city_counts = df[col_city].value_counts().head(15)
    plt.figure(figsize=(10, 6))
    sns.barplot(x=city_counts.values, y=city_counts.index)
    plt.xlabel("Number of NbS interventions")
    plt.ylabel("City")
    plt.title("Top 15 Cities by Number of NbS Interventions")
    plt.tight_layout()
    plt.show()

if col_spatial_scale in df.columns:
    plt.figure(figsize=(8, 5))
    sns.countplot(data=df, x=col_spatial_scale, order=df[col_spatial_scale].value_counts().index)
    plt.xticks(rotation=45, ha="right")
    plt.title("Distribution of Spatial Scale")
    plt.xlabel("Spatial scale")
    plt.ylabel("Count")
    plt.tight_layout()
    plt.show()

if col_area_before in df.columns:
    area_counts = df[col_area_before].value_counts().head(15)
    plt.figure(figsize=(10, 6))
    sns.barplot(x=area_counts.values, y=area_counts.index)
    plt.xlabel("Number of NbS interventions")
    plt.ylabel("Type of area before implementation")
    plt.title("Top 15 Types of Area Before Implementation")
    plt.tight_layout()
    plt.show()

if col_stage in df.columns:
    plt.figure(figsize=(8, 5))
    sns.countplot(
        data=df,
        x=col_stage,
        order=df[col_stage].value_counts().index
    )
    plt.xticks(rotation=45, ha="right")
    plt.xlabel("Present stage")
    plt.ylabel("Count")
    plt.title("Distribution of Present Stage of Intervention")
    plt.tight_layout()
    plt.show()


# Numerical variable distribution

if col_population in df.columns:
    pop = df[col_population].dropna()
    pop = pop[pop > 0]

    plt.figure(figsize=(10, 6))
    sns.histplot(pop, bins=30)
    plt.xlabel("City population")
    plt.ylabel("Frequency")
    plt.title("Distribution of City Population")
    plt.tight_layout()
    plt.show()


    plt.figure(figsize=(10, 6))
    sns.histplot(np.log10(pop), bins=30)
    plt.xlabel("log10(City population)")
    plt.ylabel("Frequency")
    plt.title("Distribution of City Population (log10 scale)")
    plt.tight_layout()
    plt.show()

# Simple geographical visualization

if col_country in df.columns:
    country_counts = df[col_country].value_counts().reset_index()
    country_counts.columns = ["Country", "Count"]

    fig = px.choropleth(
        country_counts,
        locations="Country",
        locationmode="country names",
        color="Count",
        color_continuous_scale="Viridis",
        title="Number of NbS Interventions per Country"
    )
    fig.show()


# Multi-label column analysis function

def analyze_multilabel_column(df, column_name, sep=";"):
    """
    For columns like "Ecosystem services provided" where a cell contains multiple labels:
    - Split into individual labels
    - Count frequencies
    - Plot a bar chart (top 20)
    """
    if column_name not in df.columns:
        print(f"Column {column_name} is not in the dataset. Please check the name.")
        return

    col_series = df[column_name].dropna().astype(str)

    # Split multi-label entries
    all_labels = []
    for cell in col_series:
        parts = [p.strip() for p in str(cell).split(sep) if p.strip() != ""]
        all_labels.extend(parts)

    if not all_labels:
        print(f"No valid labels found in column {column_name}.")
        return

    label_counts = pd.Series(all_labels).value_counts()

    print(f"\nLabel frequencies in column {column_name} (Top 20):")
    display(label_counts.head(20))

    # Plot bar chart
    top_n = 20
    plt.figure(figsize=(10, 6))
    sns.barplot(
        x=label_counts.head(top_n).values,
        y=label_counts.head(top_n).index
    )
    plt.xlabel("Count")
    plt.ylabel("Label")
    plt.title(f"Top {top_n} labels in '{column_name}'")
    plt.tight_layout()
    plt.show()

# analyze_multilabel_column(df, "Ecosystem services provided", sep=";")
# analyze_multilabel_column(df, "Social impacts", sep=";")
# analyze_multilabel_column(df, "Environmental impacts", sep=";")
```
